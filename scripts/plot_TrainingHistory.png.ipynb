```python
import numpy as np
import pickle

import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
matplotlib.rcParams.update({'font.size': 18})
```

```python
VAEversion = '71'
```

```python
f = open('../VAE/VAE_HLFONLY/TrainingHistory_VAE_all-in-one_v'+VAEversion+'.pkl', 'r')
history = pickle.load(f)
f.close()
```

```python
f = plt.figure(figsize=(16,8))
style = {'loss':'--', 'RecoNLL_loss': 'o', 'KL_loss': '--'}

labels = {'loss':r'$Loss_{TOT}$', 'RecoNLL_loss':r'$Loss_{RECO}$', 'KL_loss':r'$D_{KL}$'}

for item in style.keys():
    plt.plot(history[item][1:], style[item], label=labels[item])
    plt.plot(history['val_'+item][1:], style[item], label='Val '+labels[item])

plt.xlabel('Epoch')
plt.grid()
plt.legend(loc='best')
plt.yscale('log')
f.savefig('../fig/paper_fig/TrainingHistory_VAEv'+ VAEversion +'.png')
```
