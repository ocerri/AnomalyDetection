```python
import numpy as np
import h5py
import matplotlib.pyplot as plt
from glob import glob
import re
import ROOT as rt
import root_numpy as rtnp
import itertools

rt.gErrorIgnoreLevel = 4000
```

```python
trigger_rate = {'Wlnu':440. ,'qcd':251., 'ttbar':2.5} #Hz

acquisition_sec_per_year = 1e7
lumi_per_yr = 50 #fb^-1
```

```python
VAE_train_sample = []
samples = []
loss = {}

for i, fpath in enumerate(glob('losses/*.h5')):
    out = re.search('/.*_v[0-9]', fpath)
    
    VAE_ts =  out.group(0)[1:-3]
    VAE_train_sample.append(VAE_ts)
    
    f = h5py.File(fpath, 'r')
    if i == 0:
        for k in f.keys():
            loss[str(k)] = []
            samples.append(str(k))
    
    for k in f.keys():
        k = str(k)
        loss[k].append(np.array(f[k]))
    
    f.close()
    
for k in loss.keys():
    loss[k] = np.array(loss[k])
```

1. First index is the sample on which VAE has been trained
2. Second index is the loss type (Tot, nll, kl)
3. Third index is the Event number

```python
print loss['qcd'].shape
colors = {'AtoChHW_HIGHMASS':1, 'qcd':7, 'Wlnu':4, 'AtoChHW':rt.kYellow+2, 'Wprime':6, 'Ato4l':2, 'Zprime':46, 'ttbar':8}
histos = []
canvases = []
legends = []
quantile_to_cut = 1e-3
Ytitles = ['Loss tot', 'nll(reco probability)', 'KL divergence']
```

```python
for k, VAE_ts in enumerate(VAE_train_sample):
    c = rt.TCanvas('c'+VAE_ts, 'c'+VAE_ts, 1200, 600)
    c.Divide(3)
    
    for loss_piece in range(3):        
        pad = c.cd(loss_piece+1)
        pad.SetLeftMargin(0.2)
        leg = rt.TLegend(0.7,0.2, 1, 1)
        leg.SetTextSize(0.04)
        loss_train = loss[VAE_ts][k, loss_piece]
        binning = [100, np.min(loss_train), np.percentile(loss_train, 100*(1-quantile_to_cut/10))]
        
        h = rt.TH1F('h_loss_{}_{}_{}'.format(VAE_ts, VAE_ts, loss_piece), VAE_ts, binning[0], binning[1], binning[2])
        rtnp.fill_hist(h, loss_train)
        h.Scale(1./loss_train.shape[0])

        h.SetStats(0)
        h.SetLineColor(colors[VAE_ts])
        h.SetFillColorAlpha(colors[VAE_ts], 0.5)
        h.SetFillStyle(3001)
        h.Draw('Bar SAME')
        histos.append(h)

        val_to_cut = np.percentile(loss_train, 100*(1-quantile_to_cut))
        leg.AddEntry(h, '{:.1f}'.format(100*quantile_to_cut), 'lf')
#         leg.AddEntry(h, h.GetTitle()+' ({:1.1e})'.format(quantile_to_cut), 'lf')
        line = rt.TLine()
        line.SetLineColor(15)
        line.SetLineWidth(3)
        line.SetLineStyle(9)
        line.DrawLine(val_to_cut,0, val_to_cut,1)
#         print '\t{} {:1.1e}'.format(VAE_ts, quantile_to_cut)

        for n in loss.keys():
            if n == VAE_ts:
                continue
                
            z = loss[n][k,loss_piece]
            eff = np.sum(z>val_to_cut)/float(z.shape[0])
#             print '\t{} {:1.1e}'.format(n, eff)
            h1 = rt.TH1F('h_loss_{}_{}_{}'.format(n, VAE_ts, loss_piece), n+' ({:1.1e})'.format(eff), binning[0], binning[1], binning[2])
            rtnp.fill_hist(h1, z)
            h1.Scale(1./float(z.shape[0]))

            h1.SetStats(0)
            h1.SetLineColor(colors[n])
            h1.SetLineWidth(2)
            h1.Draw('SAME')
            leg.AddEntry(h1, '{}{} {:.1f}'.format(n[0],n[-2:], 100*eff), 'l')
#             leg.AddEntry(h1, h1.GetTitle(), 'l')
            histos.append(h1)
        
        leg.Draw()
        legends.append(leg)
        h.SetTitle('')
        h.SetXTitle(Ytitles[loss_piece])
        h.SetYTitle('Probability')
        pad.SetGrid()
        pad.SetLogy()
        
    c.Draw()
    canvases.append(c)
```

```python
from prettytable import PrettyTable
```

```python
loss_to_be_used = 2
# cuts_on_loss = np.array([31, 30, 30])
# cuts_on_loss = np.array([31, 24, 30])
# cuts_on_loss = np.array([24, 30, 27])
cuts_on_loss = np.array([31, 20, 44])

table = PrettyTable(['Sample', 'Efficiency','Rate [Hz]', 'evts/month'])
N_tot = 0
d_N_tot2 = 0
for n, lss in loss.items():
    if n in VAE_train_sample:
        aux = np.greater(lss[:,2,:].T, cuts_on_loss)
        aux = aux.T
        selected = np.sum(np.logical_and(np.logical_and(aux[0], aux[1]), aux[2]))
        eff = float(selected)/lss.shape[2]
        d_eff = np.sqrt(eff*(1-eff)/lss.shape[2])
        rate = trigger_rate[n]*eff
        N_evts = rate*acquisition_sec_per_year/12.
        N_tot += N_evts
        d_N_evts = trigger_rate[n]*d_eff*acquisition_sec_per_year/12.
        d_N_tot2 += np.square(d_N_evts)
        table.add_row([n, '{:1.1e} +/- {:1.1e}'.format(eff, d_eff), '{:1.1e}'.format(rate), '{:.0f} +/- {:.0f}'.format(N_evts, d_N_evts)])
        
print table

print 'Expected evts/month: {:.0f} +/- {:.0f}'.format(N_tot, np.sqrt(d_N_tot2))
print 

table = PrettyTable(['Sample', 'Efficiency', 'xsec (10 evts/month) [fb]', 'xsec (S/B = 0.3) [fb]'])

for n, lss in loss.items():
    if not (n in VAE_train_sample):
        aux = np.greater(lss[:,2,:].T, cuts_on_loss)
        aux = aux.T
        selected = np.sum(np.logical_and(np.logical_and(aux[0], aux[1]), aux[2]))
        eff = float(selected)/lss.shape[2]
        d_eff = np.sqrt(eff*(1-eff)/lss.shape[2])
        x_sec_lim = 10. * 12/ (lumi_per_yr * eff)
        x_sec_SoB = 0.3*N_tot*12/ (lumi_per_yr * eff)
        table.add_row([n, '{:1.1e} +/- {:1.1e}'.format(eff, d_eff), '{:.1f}'.format(x_sec_lim), '{:1.1e}'.format(x_sec_SoB)])
        
print table
```

```python
def AverageBest3BkgOSig_ratio(cuts_on_loss, loss_to_be_used=2):
    N_bkg = 0
    for n, lss in loss.items():
        if n in VAE_train_sample:
            aux = np.greater(lss[:,2,:].T, cuts_on_loss)
            aux = aux.T
            selected = np.sum(np.logical_and(np.logical_and(aux[0], aux[1]), aux[2]))
            eff = float(selected)/lss.shape[2]
            rate = trigger_rate[n]*eff
            N_bkg += rate*acquisition_sec_per_year/12

    x_sec_lim = []
    for n, lss in loss.items():
        if not (n in VAE_train_sample):
            aux = np.greater(lss[:,2,:].T, cuts_on_loss)
            aux = aux.T
            selected = np.sum(np.logical_and(np.logical_and(aux[0], aux[1]), aux[2]))
            eff = float(selected)/lss.shape[2]
            x_sec_lim.append( 10. * 12/ (lumi_per_yr * eff) )
            
    x_sec_lim.sort()
#     print x_sec_lim
#     print float(N_bkg)
#     print np.mean(x_sec_lim[:3])
    return float(N_bkg) + 10*np.mean(x_sec_lim[:3])
```

```python
# aux = cuts_on_loss
# val_prec = AverageBest3BkgOSig_ratio(cuts_on_loss)
# val_curr = AverageBest3BkgOSig_ratio(cuts_on_loss)

directions = np.array([[0,0,0], [0,0,1], [0,1,0], [1,0,0], [0,0,-1], [0,-1,0], [-1,0,0]])

N = 0
aux = np.array([25,25,25])
while(True):
    if N > 100:
        print 'Max Iteration reached'
        break

    vals = []
    for x in aux + directions:
        vals.append(AverageBest3BkgOSig_ratio(x))
    
    i_min = np.argmin(vals)
    
    N += 1
    
    if i_min == 0:
        print 'minimum found'
        break
    else:
        aux = (aux + directions)[i_min]
        print 'New minimum {:.1f} in {}'.format(vals[i_min], aux)
```

```python
val_min = AverageBest3BkgOSig_ratio(cuts_on_loss)
x_min = cuts_on_loss

N = 0
for a,b,c in itertools.product(range(15,45), range(20, 80), range(15,45)):
    N +=1
    if N%1000 == 0:
        print N
    aux_x = np.array([a,b,c])
    aux_val = AverageBest3BkgOSig_ratio(aux_x)
    if aux_val < val_min:
        val_min = aux_val
        x_min = aux_x
        print 'New minimum {:.1f} in {}'.format(val_min, x_min)
```
